```python
import pandas as pd 
df = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv')
df2 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv')
df3 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv')
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
df2.columns = df2.columns.str.lower().str.replace(' ', '_')
df3.columns = df3.columns.str.lower().str.replace(' ', '_')

df = df.rename(columns={'st': 'state'})
df2 = df2.rename(columns={'st': 'state'})

print(df.columns)
print(df2.columns)
print(df3.columns)
```

```python
# Create sets of columns for each DataFrame
cols_df1 = set(df.columns)
cols_df2 = set(df2.columns)
cols_df3 = set(df3.columns)

# Check if all are equal
if cols_df1 == cols_df2 == cols_df3:
    print("The three DataFrames have the same columns.")
else:
    print("The DataFrames have different columns.")

# Optional: Show the differences
print("Columns in df1 but not in df2:", cols_df1 - cols_df2)
print("Columns in df2 but not in df1:", cols_df2 - cols_df1)
print("Columns in df3 but not in df1:", cols_df3 - cols_df1)
```

```python
customers_df = set(df["customer"])
customers_df2 = set(df2["customer"])
customers_df3 = set(df3["customer"])

print("Clientes en df y df2:", len(customers_df & customers_df2))
print("Clientes en df y df3:", len(customers_df & customers_df3))
print("Clientes en df2 y df3:", len(customers_df2 & customers_df3))
print("Clientes en todos los DataFrames:", len(customers_df & customers_df2 & customers_df3))
```

```python
df_combined = pd.concat([df, df2, df3], ignore_index=True)
```

```python
df_combined
```

```python
df_combined = df_combined.dropna(subset=['customer'])
```

```python
duplicates = df_combined[df_combined.duplicated(subset='customer')]

print(duplicates)
```

```python
df_combined = df_combined.drop_duplicates(keep='first')
```

```python
df_combined[df_combined.duplicated(subset='customer')]
```

```python
df_combined[df_combined['customer'] == 'AC40767']
```

```python
pd.set_option('display.max_rows', None)  # Exibe todas as linhas

df_duplicates = df_combined[df_combined.duplicated(subset='customer', keep=False)]

df_duplicates_sorted = df_duplicates.sort_values(by='customer')

df_duplicates_sorted
```

```python
df_combined = df_combined.drop([856, 4026, 4616, 4283, 354, 4124, 741])
```

```python
def normalize_clv(clv):
    if isinstance(clv, str): 
        if '%' in clv:  
            clv = clv.replace('%', '')  
            return float(clv) / 100  
        else:
            return float(clv)  
    else:
        return clv  

df_combined['customer_lifetime_value'] = df_combined['customer_lifetime_value'].apply(normalize_clv)

print(df_combined['customer_lifetime_value'].head())
```

```python
df_combined = df_combined.drop_duplicates(subset='customer', keep='first')
```

```python
df_combined[df_combined.duplicated(subset='customer')]
```

```python
df_combined['gender'].unique()
```

```python
df_combined['gender'] = df_combined['gender'].replace({
    'Male': 'M',
    'female': 'F',
    'Femal': 'F'
})
```

```python
df_combined['state'].unique()
```

```python
df_combined['state'] = df_combined['state'].replace({"AZ": "Arizona", "Cali": "California", "WA": "Washington"})
```

```python
df_combined['education'].unique()
```

```python
df_combined['education'] = df_combined['education'].replace({'Bachelors': 'Bachelor'})
```

```python
df_combined['vehicle_class'].unique()
```

```python
df_combined['vehicle_class'] = df_combined['vehicle_class'].replace(['Sports Car', 'Luxury SUV', 'Luxury Car'], 'Luxury')
```

```python
df_combined.number_of_open_complaints.unique()
```

```python
df_combined['number_of_open_complaints'] = df_combined['number_of_open_complaints'].str.split('/').str[1].astype(float)
```

```python
df_combined['number_of_open_complaints'] = df_combined['number_of_open_complaints'].astype('Int64')
```

```python
df_combined.number_of_open_complaints.unique()
```

```python
df_combined.dtypes
```

```python
df_combined.isnull().sum()
```

```python
mean_clv = df_combined['customer_lifetime_value'].mean()

df_combined['customer_lifetime_value'] = df_combined['customer_lifetime_value'].fillna(mean_clv)

df_combined.isnull().sum()
```

```python
df_combined['gender'] = df_combined['gender'].fillna('Unknown')
df_combined['number_of_open_complaints'] = df_combined['customer_lifetime_value'].fillna(0)

df_combined.isnull().sum()
```

```python
df_combined.duplicated().sum()
```

```python
df_combined = df_combined.reset_index()
```

1. You work at the marketing department and you want to know which sales channel brought the most sales in terms of total revenue. Using pivot, create a summary table showing the total revenue for each sales channel (branch, call center, web, and mail).
Round the total revenue to 2 decimal points.  Analyze the resulting table to draw insights.

```python
revenue_by_channel = mkt.pivot_table(index='sales_channel', values='monthly_premium_auto', aggfunc='sum')
revenue_by_channel = revenue_by_channel.round(2)
revenue_by_channel
```

2. Create a pivot table that shows the average customer lifetime value per gender and education level. Analyze the resulting table to draw insights.

```python
avg_clv = mkt.pivot_table(index='gender', columns='education', values=['customer_lifetime_value'], aggfunc='mean')
avg_clv
```

Show it in a long format table.

```python
mkt.columns
```

*In data analysis, a long format table is a way of structuring data in which each observation or measurement is stored in a separate row of the table. The key characteristic of a long format table is that each column represents a single variable, and each row represents a single observation of that variable.*

*More information about long and wide format tables here: https://www.statology.org/long-vs-wide-data/*

```python
complaints_summary = mkt.groupby(['month', 'policy_type'])['number_of_open_complaints'].sum().reset_index()
```
